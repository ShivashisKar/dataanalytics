```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default = "plotly_white"
```

```python
# Assuming 'venue' is the column you want to group by
# And assuming there are other columns in the DataFrame with numerical data
numeric_columns = df_cricket_dataset_copy.select_dtypes(include=[np.number]).columns.tolist()
# Group by 'venue' and sum only the numeric columns
result = df_cricket_dataset_copy.groupby('venue')[numeric_columns].sum()
result["first innings score"]
```

```python
x_axis_1=np.arange(df_cricket_dataset_copy["venue"].nunique())
x_axis_1
```

```python
plt.figure(figsize=(12,8))
plt.bar(x_axis_1-0.2,result["first innings score"],width=0.4)
plt.bar(x_axis_1+0.2,result["second innings score"],width=0.4)

plt.xticks(x_axis_1,pd.unique(df_cricket_dataset_copy["venue"].sort_values()))

plt.xlabel("Venue",fontdict={"fontsize":14,
                            "fontweight":"bold"})

plt.ylabel("Scores",fontdict={"fontsize":14,
                                        "fontweight":"bold"})

plt.legend(["First Innings Score","Second Innings Score"])

plt.title("First VS Second Innings Score",fontdict={"fontsize":20,
                                        "fontweight":"bold"})
```
